```python
data1 = Zee.tot_layers_et() / Zee.et

mask = data1 < 2.0
data1 = data1[mask]

# Define the number of bins
num_bins = 150
bin_edges = np.linspace(np.min(data1), np.max(data1), num_bins + 1)

# Calculate histograms manually and normalize by the total number of occurrences
hist1, _ = np.histogram(data1, bins=bin_edges)

total_occurrences1 = len(data1)

hist_normalized1 = hist1 / total_occurrences1


with plt.style.context(["science", "notebook", "grid"]):
    plt.figure(figsize=(12, 8))
    plt.hist(
        bin_edges[:-1],
        bins=bin_edges,
        weights=hist_normalized1,
        histtype="step",
        label="Z to ee data",
        # linestyle="dashed",
        color="k",
    )

    # Set labels and title
    plt.xlabel("Fraction of particle's energy deposited in calorimeters")
    plt.ylabel("Frequency")
    plt.title("Total energy in calorimeters / true et")

    # Add legend
    plt.legend()

    # Show plot
    plt.show()
```

```python
hist, bins = np.histogram(Zee.z, bins=100, density=True)
bin_centers = 0.5 * (bins[:-1] + bins[1:])

initial_guess = [1400, 0, 40]

# Perform curve fitting
popt, pcov = curve_fit(utils.gaussian, bin_centers, hist, p0=initial_guess)

# Extract fitted parameters and their uncertainties
A_fit, mu_fit, std_fit = popt
A_err, mu_err, std_err = np.sqrt(np.diag(pcov))

print(mu_fit, mu_err)
print(std_fit, std_err)
```

```python
with plt.style.context(["science", "notebook", "grid"]):
    plt.figure(figsize=(8, 6), dpi=150)
    n, bins, _ = plt.hist(
        Zee.z,
        bins=100,
        histtype="step",
        label=r"$Z \rightarrow e^+e^-$ data",
        color="k",
        density=False,
    )

    # Plot the Gaussian fit scaled to occurrences
    bin_centers = 0.5 * (bins[:-1] + bins[1:])
    bin_width = bins[1] - bins[0]
    scaling_factor = len(Zee.z) * bin_width
    plt.plot(
        bin_centers,
        scaling_factor * utils.gaussian(bin_centers, A_fit, mu_fit, std_fit),
        "r",
        linewidth=1,
        label="Gaussian fit",
    )

    # Add fitted mean and standard deviation as text
    plt.text(
        0.33,
        0.96,
        f"Mean: ({mu_fit:.2f} $\pm$ {mu_err:.2f}) mm \n Stddev: ({std_fit:.1f} $\pm$ {std_err:.1f}) mm",
        ha="right",
        va="top",
        transform=plt.gca().transAxes,
        bbox=dict(facecolor="white", alpha=0.8),
    )

    plt.xlabel("z [mm]")
    plt.ylabel("Occurrences")
    plt.title(r"$Z \rightarrow e^+e^-$")
    plt.legend()
    plt.show()
```
